```python
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    rolling_mean = timeseries.rolling(window=12).mean()
    rolling_std = timeseries.rolling(window=12).std()
    
    #Plot rolling statistics
    plt.figure(figsize=(17,8))
    plt.title('Rolling Mean and Standard Deviation')
    plt.plot(timeseries, 'blue', label='Actual values')
    plt.plot(rolling_mean, 'green', label='Rolling mean')
    plt.plot(rolling_std, 'black', label='Rolling std')
    plt.legend(loc='best')
    plt.grid(True)
    
    #Perform Dickey-Fuller test
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for key, value in dftest[4].items():
        dfoutput[f'Critical Value {key}'] = value
    print(dfoutput)
```

```python
test_stationarity(main_df['TOTAL_HOME_BURGLARIES'])
```

ADF test statistic value < Critical value --> reject null hypothesis --> Time Series is stationary
although the mean is changing during the time, the ADF test "thinks" that the series looks more like a stationary AR than like a random walk. And here I agree with it.

```python
from scipy.optimize import minimize
import statsmodels.tsa.api as smt

def tsplot(y, lags=None, figsize=(12, 7)):
    fig = plt.figure(figsize=figsize)
    layout = (1,2)
    acf_ax = plt.subplot2grid(layout, (0,0))
    pacf_ax = plt.subplot2grid(layout, (0,1))

    smt.graphics.plot_acf(y, lags=lags, ax=acf_ax)
    smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax)
    plt.tight_layout()
        
tsplot(main_df, lags=30)

# Take the first difference to remove to make the process stationary
```

From the ACF plot above, we can see that our seasonal period consists of 12 timesteps (months) (where the ACF has the second largest positive peak and in the 95% confidence interval).

So, because there is seasonality in our data, I will use SARIMA and try to predict the future :)

p and P somewhere between 0 and 2 (significant positive spikes in ACF and PACF plots)

d and D both equal 1 (first difference and seasonal difference)

q should be 0 (there aren't significant negative spikes in the ACF)
and Q may be 5

s equal to 12

```python
#Set initial values and some bounds
ps = range(3)
d = 1
qs = range(3)
Ps = range(3)
D = 1
Qs = [0,5]
s = 12

#Create a list with all possible combinations of parameters
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

```python
import warnings
warnings.filterwarnings('ignore')

def optimize_SARIMA(parameters_list, d, D, s):
    """
        Return dataframe with parameters and corresponding AIC
        
        parameters_list - list with (p, q, P, Q) tuples
        d - integration order
        D - seasonal integration order
        s - length of season
    """
    
    results = []
    best_aic = float('inf')
    
    for param in tqdm_notebook(parameters_list):
        try: model = sm.tsa.statespace.SARIMAX(main_df, order=(param[0], d, param[1]),
                                               seasonal_order=(param[2], D, param[3], s)).fit(disp=-1)
        except Exception as e:
            print(e)
            continue
            
        results.append([param, model.aic])
        
    result_table = pd.DataFrame(results)
    print(result_table)
    result_table.columns = ['parameters', 'aic']
    #Sort in ascending order, lower AIC is better
    result_table = result_table.sort_values(by='aic', ascending=True).reset_index(drop=True)
    
    return result_table

result_table = optimize_SARIMA(parameters_list, d, D, s)
```

There is not enough samples to try all of the combinations we wanted (for example  Q=5 wasnt tested at all)

```python
#Set parameters that give the lowest AIC (Akaike Information Criteria)

p, q, P, Q = result_table.parameters[0]

best_model = sm.tsa.statespace.SARIMAX(main_df, order=(p, d, q),
                                       seasonal_order=(P, D, Q, s)).fit(disp=-1)

print(best_model.summary())
```

```python
# load libraries
import math
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
import matplotlib.pyplot as plt
```

```python
# "load" the dataset
dataset = df_hb.groupby(['YEAR', 'MONTH'])['CMPLNT_NUM'].count().to_frame().reset_index()
dataset = dataset[dataset['YEAR'].apply(int).isin(range(2006, 2019))]['CMPLNT_NUM'].values
dataset = dataset.reshape(-1, 1)

# normalize the dataset so all of the data will be values between 0 and 1 
scaler = MinMaxScaler(feature_range=(0, 1)) 
scaled_data = scaler.fit_transform(dataset)

# Split the dataset into training and test sets
train_data = scaled_data[:144] #untill the end of 2017
test_data =  scaled_data[-24:] #from the beggining of 2016 because we are going to use 24 months steps in order to predict the 25th month (januar 2018).s\
```

```python
x_train = []
y_train = []
for i in range(12,len(train_data)):
    x_train.append(train_data[i-12:i,0])
    y_train.append(train_data[i,0])
    
#Convert x_train and y_train to numpy arrays
x_train, y_train = np.array(x_train), np.array(y_train)

#Reshape the data into the shape accepted by the LSTM ([number of samples, number of time steps, and number of features])
x_train = np.reshape(x_train, (x_train.shape[0],x_train.shape[1],1))
```

```python
#Build the LSTM network model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True,input_shape=(x_train.shape[1],1)))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dense(units=25))
model.add(Dense(units=1))

#Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

#Train the model
model.fit(x_train, y_train, batch_size=1, epochs=30)
```

```python
#Create the x_test and y_test data sets
x_test = []
y_test =  dataset[-12 : , : ] #get all the labels for 2018 (unscaled!)
for i in range(12,len(test_data)):
    x_test.append(test_data[i-12:i,0])
    
#Convert x_test to a numpy array 
x_test = np.array(x_test)

#Reshape the data into the shape accepted by the LSTM
x_test = np.reshape(x_test, (x_test.shape[0],x_test.shape[1],1))
```

```python
#Getting the models predicted no. home burglaries values
predictions = model.predict(x_test) 
predictions = scaler.inverse_transform(predictions)#Undo scaling
```

```python
#Calculate/Get the value of RMSE
rmse=np.sqrt(np.mean(((predictions- y_test)**2)))
rmse
```

```python
# Lets see how our model is doing on the training set, and out of curiosity, for the rest of 2018 as well
plt.figure(figsize=(16,9))
plt.plot(predictions, 'blue', label='predictions')
plt.plot(testset[:12].values, 'green', label='real future values')
plt.legend(loc='best')
```
